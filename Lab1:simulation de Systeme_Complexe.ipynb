This helper function replaces the simulated LLM function.

```python
def prompt_llm(prompt, model=MODEL_NAME, max_output_tokens=400):
    response = client.responses.create(
        model=model,
        input=prompt,
        max_output_tokens=max_output_tokens
    )
    return response.output_text
```

# Part 1: Prompt Chaining

**Prompt chaining** breaks a task into ordered steps, where each output becomes the input to the next step.

```python
topic = "Artificial Intelligence in Education"

outline = prompt_llm(f"Create a short outline about: {topic}")
draft = prompt_llm(f"Expand this outline into a short report:\n{outline}")
summary = prompt_llm(f"Summarize this report in 5 lines:\n{draft}")

print("=== OUTLINE ===")
print(outline)

print("\n=== DRAFT ===")
print(draft)

print("\n=== SUMMARY ===")
print(summary)
```

# Part 2: Routing

**Routing** chooses a processing path based on the user’s request type.

```python
def route_task(user_input):
    text = user_input.lower()

    if "translate" in text:
        return "translation"
    elif "summarize" in text or "summary" in text:
        return "summarization"
    elif "code" in text or "python" in text:
        return "code_generation"
    else:
        return "general_qa"

def handle_task(user_input):
    task_type = route_task(user_input)

    if task_type == "translation":
        return prompt_llm(f"Translate the following text:\n{user_input}")
    elif task_type == "summarization":
        return prompt_llm(f"Summarize the following text:\n{user_input}")
    elif task_type == "code_generation":
        return prompt_llm(f"Write Python code for the following request:\n{user_input}")
    else:
        return prompt_llm(f"Answer the following question:\n{user_input}")
```

```python
queries = [
    "Translate this sentence into French: Artificial intelligence is useful.",
    "Summarize the importance of deep learning.",
    "Write Python code to compute factorial.",
    "What is reinforcement learning?"
]

for q in queries:
    print("USER:", q)
    print("ROUTE:", route_task(q))
    print("OUTPUT:", handle_task(q))
    print("-" * 60)
```

# Part 3: Parallelization

**Parallelization** runs independent subtasks at the same time.

```python
from concurrent.futures import ThreadPoolExecutor

document = """
Artificial intelligence is transforming education, healthcare, transportation,
and industry. It enables automation, prediction, and decision support.
"""

def get_summary():
    return prompt_llm(f"Summarize the following text:\n{document}")

def get_keywords():
    return prompt_llm(f"Extract keywords from the following text:\n{document}")

def get_questions():
    return prompt_llm(f"Generate 3 discussion questions from the following text:\n{document}")

with ThreadPoolExecutor() as executor:
    future_summary = executor.submit(get_summary)
    future_keywords = executor.submit(get_keywords)
    future_questions = executor.submit(get_questions)

summary = future_summary.result()
keywords = future_keywords.result()
questions = future_questions.result()

print("=== SUMMARY ===")
print(summary)

print("\n=== KEYWORDS ===")
print(keywords)

print("\n=== QUESTIONS ===")
print(questions)
```

## Exercise 3

Use parallelization to analyze a text from three angles:
- summary,
- important concepts,
- possible applications.

```python
text = """
Machine learning is widely used in medical diagnosis, fraud detection,
recommendation systems, and self-driving cars.
"""

def get_summary():
    return prompt_llm(f"Summarize this text:\n{text}")

def get_concepts():
    return prompt_llm(f"Extract the most important concepts from this text:\n{text}")

def get_applications():
    return prompt_llm(f"List practical applications mentioned or implied in this text:\n{text}")

with ThreadPoolExecutor() as executor:
    f1 = executor.submit(get_summary)
    f2 = executor.submit(get_concepts)
    f3 = executor.submit(get_applications)

print("=== SUMMARY ===")
print(f1.result())

print("\n=== CONCEPTS ===")
print(f2.result())

print("\n=== APPLICATIONS ===")
print(f3.result())
```

# Part 4: Reflection

**Reflection** adds a second pass: first answer, then critique, then improve.

```python
question = "Explain overfitting in machine learning."

initial_answer = prompt_llm(f"Answer clearly:\n{question}")
critique = prompt_llm(f"Review this answer and identify weaknesses:\n{initial_answer}")
improved_answer = prompt_llm(
    f"Improve the answer below using this critique.\n\n"
    f"Answer:\n{initial_answer}\n\nCritique:\n{critique}"
)

print("=== INITIAL ANSWER ===")
print(initial_answer)

print("\n=== CRITIQUE ===")
print(critique)

print("\n=== IMPROVED ANSWER ===")
print(improved_answer)
```

# Part 5: Tool Use

**Tool use** lets the workflow call external functions.

```python
def calculator(expression):
    try:
        return eval(expression)
    except Exception as e:
        return f"Error: {e}"

def tool_use_system(user_input):
    if any(op in user_input for op in ["+", "-", "*", "/"]):
        return f"Calculated result: {calculator(user_input)}"
    else:
        return prompt_llm(f"Respond to this request:\n{user_input}")
```

```python
print(tool_use_system("25 * 4 + 10"))
print(tool_use_system("Explain what supervised learning is"))
```

# Part 7: Multi-Agent Systems

A **multi-agent system** uses several role-based prompts that collaborate on one task.

```python
def agent_planner(topic):
    return prompt_llm(
        f"You are a planner agent. Create a structured plan for a report about: {topic}"
    )

def agent_writer(plan):
    return prompt_llm(
        f"You are a writer agent. Write a report based on this plan:\n{plan}"
    )

def agent_reviewer(draft):
    return prompt_llm(
        f"You are a reviewer agent. Improve the following report:\n{draft}"
    )
```

```python
topic = "Applications of AI in Healthcare"

plan = agent_planner(topic)
draft = agent_writer(plan)
final_version = agent_reviewer(draft)

print("=== PLAN ===")
print(plan)

print("\n=== DRAFT ===")
print(draft)

print("\n=== FINAL VERSION ===")
print(final_version)
```

# Final Mini Project

Build a mini assistant that:
1. routes the user request,
2. uses a tool when needed,
3. uses planning for code tasks,
4. uses reflection before returning the final answer.

```python
def intelligent_assistant(user_input):
    route = route_task_extended(user_input)

    if any(op in user_input for op in ["+", "-", "*", "/"]):
        return f"Tool Result: {calculator(user_input)}"

    if route == "summarization":
        first = prompt_llm(f"Summarize this text:\n{user_input}")
        critique = prompt_llm(f"Review this summary and suggest improvements:\n{first}")
        final = prompt_llm(
            f"Improve this summary using the critique:\nSummary: {first}\nCritique: {critique}"
        )
        return final

    elif route == "translation":
        return prompt_llm(f"Translate the following:\n{user_input}")

    elif route == "code_generation":
        plan = prompt_llm(f"Plan the code solution for this request:\n{user_input}")
        code = prompt_llm(f"Write Python code based on this plan:\n{plan}")
        return code

    elif route == "email_writing":
        draft = prompt_llm(f"Write a professional email for:\n{user_input}")
        critique = prompt_llm(f"Review and improve this email:\n{draft}")
        improved = prompt_llm(
            f"Generate the final improved email:\nDraft: {draft}\nCritique: {critique}"
        )
        return improved

    else:
        answer = prompt_llm(f"Answer the following question:\n{user_input}")
        critique = prompt_llm(f"Review this answer and identify weaknesses:\n{answer}")
        improved = prompt_llm(
            f"Improve this answer:\nAnswer: {answer}\nCritique: {critique}"
        )
        return improved
```

```python
tests = [
    "Summarize the role of AI in education.",
    "25 + 90 / 3",
    "Write Python code to compute Fibonacci numbers.",
    "Write an email to my professor asking for a meeting.",
    "What is deep learning?"
]

for t in tests:
    print("USER INPUT:", t)
    print("ASSISTANT OUTPUT:", intelligent_assistant(t))
    print("=" * 70)
```
